#**Уровень 0:**
**Задание 1** \
Посчитайте частные производные функций: \
1) $f(x,y)=2x^2y^3 + 1/x + y^2x + 7$ \
2) $f(x,y)=x^2y - sin(xy) + cos(x^2) + 6y$


```python
from sympy import *
```

```python
x,y = symbols('x,y')
```

```python
# first example
res = (2 * x**2 * y**3 + 1) / (x + y**2 * x + 7)
res
```

```python
res1 = diff(res,x)
res1
```

```python
res2 = diff(res,y)
res2
```

```python
#second example
res_1 = x**2*y - sin(x*y)+cos(x**2)+6*y
res_1
```

```python
diff(res_1,x)
```

```python
diff(res_1,y)
```

**Задание 2** \
Вычислите градиент функции: \
$f(x,y)=2x^2y^3 + 1/x + y^2x + 7$ \
в точке $(1,2)$

```python
zx = diff((2 * x**2 * y**3 + 1) / (x + y**2 * x + 7),x)
zx
```

```python
zy = diff((2 * x**2 * y**3 + 1) / (x + y**2 * x + 7),y)
zy
```

```python
round(zx.evalf(subs={x:1,y:2}),4),round(zy.evalf(subs={x:1,y:2}),4)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
x = np.linspace(-4*np.pi,4*np.pi,50)

# create data points
x = np.linspace(-10, 10, 100)
y = np.linspace(-15, 15, 100)
# create grid
X, Y = np.meshgrid(x, y)
Z = Y**2 - Y*X**2 + 2 * X * Y
# create figure container
fig = plt.figure(figsize=(9, 6))
ax = plt.axes(projection = '3d')
# 3d contour plot
ax.contour3D(X, Y, Z, 100, cmap = 'viridis')
# set labels
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
# save figure
plt.savefig('3d_contour.png', dpi = 300)
plt.show()
```
