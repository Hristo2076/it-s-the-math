**Задание 1** \
Посчитайте значение первой и второй производной функции: \
$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ \
В точке x = 1.

```python
import numpy as np
from sympy import diff, cos, sin,Symbol
from matplotlib import pyplot as plt
from sympy.solvers.solveset import linsolve
```

```python
x = Symbol('x')
```

```python
y = x**5 + 4 * sin(2*x) + cos(3*x+3)
y_dif = y.diff(x)
y_dif_2 = y_dif.diff(x)

print(f'First derivative: {y_dif}')
print(f'Second derivative: {y_dif_2}')


res_1 = diff(y).evalf(subs={x:1})
res_2 =diff(diff(y)).evalf(subs={x:1})

print(f'resoult 1: {res_1} \nresoult 2: {res_2}')
```

Постройте график первой и второй производной функции: \
$f(x) = sin^5(2x+1)$ \
На отрезке [-5; 5].

```python
print(diff(sin(2*x+1)**5))
print(diff(diff(sin(2*x+1)**5)))
```

```python
def f1(x):
    return 10*sin(2*x + 1)**4*cos(2*x + 1)
def f2(x):
    return -20*sin(2*x + 1)**5 + 80*sin(2*x + 1)**3*cos(2*x + 1)**2
```

```python
x1 = np.linspace(-5, 5, 100)
diff1 = [f1(x1[i]) for i in range(len(x1))]
diff2 = [f2(x1[i]) for i in range(len(x1))] 
```

```python
plt.plot(x1,diff1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim([-5, 10])
plt.xlim([-5, 5]) 
plt.show()
```

```python
plt.plot(x1,diff2)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim([-20, 20])
plt.xlim([-5, 5]) 
plt.show()
```
