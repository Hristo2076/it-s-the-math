#**Задание**
При фиксированном seed=42 поэкспериментируйте с параметрами алгоритма differential_evolution: strategy, popsize, mutation.

Постройте графики количества итераций (nit) оптимизации функции ackley от значения параметра.

```python
import numpy as np
from scipy.optimize import differential_evolution
import matplotlib.pyplot as plt

%matplotlib inline
     
```

```python
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e

bounds = [(-10, 10), (-10, 10)]

result = differential_evolution(ackley, bounds,
                                strategy= 'randtobest1exp', 
                                popsize=2,
                                mutation=(0,1.5),
                                seed=42)
result
     
```

```python
plt.figure(figsize=(10,5)) 
plt.xlabel('popsize')
plt.ylabel('nit')

for _ in range(1,30):

  result = differential_evolution(ackley, bounds,
                                strategy= 'randtobest1exp', 
                                popsize=_,
                                mutation=(0,1.5),
                                tol=0.2,
                                seed=42)
  plt.scatter(_, result.nit)

plt.show()
```

```python
plt.figure(figsize=(10,5)) 
plt.xlabel('popsize')
plt.ylabel('nit')

for _ in range(1,30):

  result = differential_evolution(ackley, bounds, 
                                  popsize=_,
                                  mutation=0.1, 
                                  recombination=1,
                                  strategy='best1bin',
                                  tol=0.5, 
                                  seed=42)

  plt.scatter(_, result.nit)

plt.show()
```

```python
plt.figure(figsize=(10,5)) 
plt.xlabel('popsize')
plt.ylabel('nit')

for _ in range(1,30):

  result = differential_evolution(ackley, bounds,
                                strategy= 'randtobest1bin', 
                                popsize=6,
                                mutation=1.2,
                                tol=0.1,
                                seed=42)
  plt.scatter(_, result.nit)

plt.show()
```
